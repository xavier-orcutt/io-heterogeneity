# Calculating RMST Pseudo-Observations

**This notebook calculates RMST pseudo-observations at 1 and 2 years for advanced non-small cell lung cancer patients with high PDL1 receiving pembrolizumab plus chemotherapy or pembrolizumab.**

```python
import sys
sys.path.append('../..')

import pandas as pd

from utils.pseudo_obs import pseudo_observations_rmst
```

```python
df = pd.read_csv('../outputs/pembrochemo_pembro_features_df.csv')
```

```python
df = df.set_index('PatientID')
```

```python
df.shape
```

```python
df = df.query('adv_diagnosis_year <= 2023')
```

```python
pseudo_rmst = pseudo_observations_rmst(
    df['duration'].values,
    df['event'].values,
    tau = 365
)

df['rmst_pseudo_1y'] = pseudo_rmst
```

```python
pseudo_rmst = pseudo_observations_rmst(
    df['duration'].values,
    df['event'].values,
    tau = 730
)

df['rmst_pseudo_2y'] = pseudo_rmst
```

```python
df = df.reset_index()
```

```python
df = df[['PatientID', 'rmst_pseudo_1y', 'rmst_pseudo_2y']]
```

```python
df.to_csv('../outputs/pseudo_obs_rmst.csv', index = False)
```
