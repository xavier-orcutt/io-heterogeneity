# Calculating RMST Pseudo-Observations

**This notebook calculates RMST pseudo-observations at 1, 2, and 3 years for metastatic clear cell RCC patients treated with first-line combination immunotherapy or single-agent antiangiogenic therapy.**

```python
import sys
sys.path.append('../..')

import pandas as pd

from utils.pseudo_obs import pseudo_observations_rmst
```

```python
df = pd.read_csv('../outputs/ioio_tki_features.csv')
```

```python
df = df.set_index('PatientID')
```

```python
df.shape
```

```python
df = df.query('met_diagnosis_year <= 2020')
```

```python
df.shape
```

```python
pseudo_rmst = pseudo_observations_rmst(
    df['duration'].values,
    df['event'].values,
    tau = 365
)

df['rmst_pseudo_1y'] = pseudo_rmst
```

```python
pseudo_rmst = pseudo_observations_rmst(
    df['duration'].values,
    df['event'].values,
    tau = 730
)

df['rmst_pseudo_2y'] = pseudo_rmst
```

```python
pseudo_rmst = pseudo_observations_rmst(
    df['duration'].values,
    df['event'].values,
    tau = 1095
)

df['rmst_pseudo_3y'] = pseudo_rmst
```

```python
df = df.reset_index()
```

```python
df = df[['PatientID', 'rmst_pseudo_2y', 'rmst_pseudo_3y']]
```

```python
df.to_csv('../outputs/pseudo_obs_rmst.csv', index = False)
```
